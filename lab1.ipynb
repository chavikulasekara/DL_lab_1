```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
np.random.rand(4,4)
```

```python
np.random.exponential(1,(4,4))
```

```python
plt.hist(np.random.exponential(1,(10000,1)), density=True, bins=100, histtype="step", color="blue", label="exponential")
plt.hist(np.random.rand(10000), density=True, bins=100, histtype="step", color="red", label="uniform")
plt.hist(np.random.randn(10000), density=True, bins=100, histtype="step", color="red", label="normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
import pandas as pd
```

```python
from ast import increment_lineno
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head(10)
```

```python
#Preprocess Dataframe
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
```

```python
#Calculate pearson correlation
corr = stats_df.corr(method='pearson')
plt.figure(figsize=(9, 8))
sns.heatmap(corr, xticklabels='auto', yticklabels='auto', annot = True)
```

```python
#Calculate spearman correlation
corr = stats_df.corr(method='spearman')
plt.figure(figsize=(9, 8))
sns.heatmap(corr, xticklabels='auto', yticklabels='auto', annot = True)
```

```python
#import the dataset
from google.colab import files
import io

uploaded = files.upload()
ds = pd.read_csv(io.BytesIO(uploaded['Au_nanoparticle_dataset.csv']))
```

```python
stats_ds = ds.loc[:, ['N_total', 'N_bulk', 'N_surface', 'R_avg']]
#stats_ds = ds[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
stats_ds.head(20)
```

```python
print('Mean Values :')
stats_ds.mean()
```

```python
print('Standard Deviation Values :')
stats_ds.std()
```

```python
print('Quartile Values :')
stats_ds.quantile([0.25,0.5,0.75])
```

```python
plt.figure(figsize=(4, 1))
plt.hist(stats_ds)
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(stats_ds)

plt.show()
```

```python
g = sns.PairGrid(stats_ds)
g.map_upper(sns.histplot) #bivariate histogram
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

plt.show()
```
